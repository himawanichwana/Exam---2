### DATA CLEANING

##### Checking for null data

```python
wine.isnull().sum()
```

Tidak terdapat data yang kosong pada setiap kolumnya

##### Renaming Columns

```python
wine.columns = ['fixed_acidity', 'volatile_acidity', 'citric_acid', 'res_sugar', 'cl', 'free_so2', 'total_so2', 'density', 'ph', 'sulphates', 'alcohol', 'quality', 'quality_cat']
```

```python
wine.columns
```

```python
wine.info()
```

### DATA PREPARATION

Replacing quality category with the new range of quality

```python
wine_qual = wine.copy()
```

```python
bins = [0,3,5,7,10]
labels = ['Bad', 'Average', 'Good', 'Very Good']
wine_qual['quality_range'] = pd.cut(x=wine_qual['quality'], bins = bins, labels = labels)
print(wine_qual[['quality_range', 'quality']].head(10))
```

```python
del wine_qual['quality_cat']
```

```python
wine_qual.columns
```

### DATA VISUALIZATION

##### 1. Correlation with heatmap

```python
wine_qual.corr()
```

```python
fig, ax = plt.subplots(figsize=(20,15))
sns.heatmap(wine_qual.corr(),cmap = 'BrBG', annot = True, ax = ax)
```

Analisa : 
1. Korelasi tertinggi didapatkan dari fixed_acidity dengan citric_acidity dengan nilai korelasi +0.67
2. korelasi tertinggi kedua adalah korelasi antara alkohol dengan quality dengan nilai 0.48
3. Korelasi lainnya memilki nilai sangat kecil antar variablenya.

##### 2. Quality range of Wine with pie chart

```python
fig, ax = plt.subplots(figsize=(15,15))

cols = ['gold', 'lightcoral', 'green', 'red']
wine_qual['quality_range'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True, colors=cols)
```

```python
cdf = wine_qual.groupby('quality_range').size().reset_index(name='quality_count').sort_values('quality_count', ascending = False)
cdf
fig, ax = plt.subplots(figsize = (10,10))
sns.barplot(x = 'quality_count', y = 'quality_range', data = cdf, palette = 'coolwarm',ax=ax)
ax.set(xlabel = 'No. of quality', ylabel = 'quality')
```

```python
wine_qual.quality_range.value_counts()
```

Analisa :
1. Jumlah wine dengan kualitas baik (nilai = 6 -7) pada dataset ini memiliki precentage tertinggi dengan nilai 52.3%
2. Hanya tedapat 0.6% kualitas wine dengan kategori yang buruk (nilai = 0-3)

##### 3. Wine Quality & Alcohol with barplot

```python
fig, ax = plt.subplots(figsize = (12,6))

sns.barplot(x='quality_range', y='alcohol', data=wine_qual, ax=ax, palette ='vlag')
plt.title('Quality and Alcohol')
```

Analisa :
1. Dari hasil visualisasi diatas dapat disimpulkan bahwa wine yang memiliki kualitas terbaik rata-rata memiliki kadar alcohol diatas 11%
2. Namun dari hasil visualisasi diatas juga menggambarkan bahwa rata2 kadar alkohol baik dalam kategori buruk, average maupun yang baik tidak terlalu memiliki perbedaan banyaknya kadar alkohol didalamnya.
